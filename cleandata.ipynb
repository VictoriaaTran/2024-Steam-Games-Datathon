```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#read the dataset
gameFile = pd.read_csv('Steam Games.csv')
```

```python
gameFile.columns
```

```python
#drop irrelevant columns
gameEdited = gameFile.drop(columns=['Header image',
                                    'About the game',
                                    'Reviews',
                                    'Website',
                                    'Support url',
                                    'Screenshots',
                                    'Movies',
                                    'Notes',
                                    'Supported languages',
                                    'Full audio languages',
                                    'Required age',
                                    'Support email',
                                    'Metacritic url',
                                    'Windows','Mac', 'Linux',
                                    'DLC count', 'Peak CCU',
                                    'User score', 'Publishers', 'Score rank'
                                    ])
gameEdited.head()
```

```python
#drop rows with null values of game name
gameEdited = gameEdited.dropna(subset=['Name'])
gameEdited
```

```python
gameEdited = gameEdited.drop(gameEdited.loc[ (gameEdited['Positive'] == 0.0) + (gameEdited['Negative'] == 0.0) & (gameEdited['Average playtime forever'] == 0.0)].index)
```

```python
#drop rows without developers
gameEdited = gameEdited.dropna(subset=['Developers'])
gameEdited.shape
gameEdited.head(5)
# gameEdited = gameEdited.to_csv('gameEdited.csv', index=False)
```
